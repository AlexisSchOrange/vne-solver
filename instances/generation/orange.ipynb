```julia
using Graphs, MetaGraphsNext
using Revise
includet("../visu.jl")
includet("io.jl")
```

```julia
filename = "../../instances/orange/oti.json"
json_graph = JSON.parsefile(filename)
```

```julia
g = MetaGraph(
    DiGraph(),
    Int,
    Dict,
    Dict,
    Dict(:name => "orange", :type => "substrate")
)

correspondance = Dict()
i_vertex = 1
for node in json_graph["nodes"]
    add_vertex!(g, i_vertex, Dict(:cap=> 1, :cost => 1))
    correspondance[node["id"]] = i_vertex
    i_vertex += 1
end

for edge in json_graph["links"]
    add_edge!(g, correspondance[edge["from"]], correspondance[edge["to"]], Dict(:cap => 5, :cost => 1))
end
```
