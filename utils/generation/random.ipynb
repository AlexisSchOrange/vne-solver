```julia
using Graphs
using MetaGraphsNext
using Revise
includet("../visu.jl")
includet("tkt.jl")
```

```julia
# substrate
n_s = 30
cap_node_max = 10
cap_node_min = 7
cap_edge_max = 20
cap_edge_min = 12
cost_max = 10
cost_min = 5
proba_edge = 0.15

s_network = MetaGraph(
    DiGraph(),
    Int,
    Dict,
    Dict,
    Dict(:name=>"sn_" * string(n_s), :type=>"substrate")
)

for node in 1:n_s
    add_vertex!(s_network, node, Dict(:cap=> rand(cap_node_min:cap_node_max), :cost => rand(cost_min:cost_max)))
end
for node in 1:n_s
    for node2 in 1:n_s
        if node!=node2
            if rand() < proba_edge
                add_edge!(s_network, node, node2, Dict(:cap=> rand(cap_edge_min:cap_edge_max), :cost => rand(cost_min:cost_max)))
            end
        end
    end
end
```

```julia
visu_graph(s_network)
```

```julia
write_s_network_to_json(s_network)
```

```julia
# virtual
n_r = 20
dem_node_max = 6
dem_node_min = 3
dem_edge_max = 6
dem_edge_min = 3
proba_edge = 0.20

v_network = MetaGraph(
    DiGraph(),
    Int,
    Dict,
    Dict,
    Dict(:name=>"vr_" * string(n_r), :type=>"virtual")
)

for node in 1:n_r
    add_vertex!(v_network, node, Dict(:dem=> rand(dem_node_min:dem_node_max)))
end
for node in 1:n_r
    for node2 in 1:n_r
        if node!=node2
            if rand() < proba_edge
                add_edge!(v_network, node, node2, Dict(:dem=> rand(dem_edge_min:dem_edge_max)))
            end
        end
    end
end
```

```julia
visu_graph(v_network)
```

```julia
write_v_network_to_json(v_network)
```
